```python
a['default'] = np.where(a['default'].str.contains("yes"), 1, 0)
a['housing'] = np.where(a['housing'].str.contains("yes"), 1, 0)
a['loan'] = np.where(a['loan'].str.contains("yes"), 1, 0)
a['y'] = np.where(a['y'].str.contains("yes"), 1, 0)
a
```

```python
X=pd.concat([a.iloc[:,0:10],a.iloc[:,11:]],axis=1)
Y=a.iloc[:,10]
```

```python
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
pred_y=classifier.predict(X)
pred_y
```

```python
classifier.predict_proba(X)
```

```python
pred_y_df=pd.DataFrame({'actual': Y,'predicted': classifier.predict(X)})
```

```python
pred_y_df['correct'] = pred_y_df['actual'] == pred_y_df['predicted']
display(pred_y_df)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,pred_y)
print (confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y,pred_y))
```

```python
classifier.predict_proba(X)[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, _ = roc_curve(Y, classifier.predict_proba(X)[:,1])

auc = roc_auc_score(Y, pred_y)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area = %0.2f)'%auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```
